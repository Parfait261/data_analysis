# Importation des packages et le donnée

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
df = pd.read_csv ("data.csv")
```

```python
# Nettoyage des donnée
```

```python
df.drop("Hire_Date", axis = 1 , inplace=True )
```

```python
#Changement des types de variables
#df["Performance_Score"] = df["Performance_Score"].astype('object')
```

```python
# Fonction pour créer la colonne Satisfaction_Level
def Add_Satisfaction_Level(score) : 
  if score < 2 :
        return 'Faible'
  elif score < 3.5 :
        return 'moyen'
  elif score >= 3.5 :
      return 'élevé'


#Appel à la fonction 
df["Satisfaction_Level"] = df["Employee_Satisfaction_Score"].apply(Add_Satisfaction_Level)
```

```python
# Verification des valeur manquante 
df.isnull().sum()
```

```python
#Modification des valeurs 'others' dans la variable Gender

df= df[df.Gender != 'Other']
df
```

## Département vs satisfaction des employés

```python
Departement = df["Department"]
Satisfaction = df["Satisfaction_Level"]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))
sns.countplot(data = df, x = Departement , hue = Satisfaction) 
plt.xlabel(' Departement')
plt.ylabel('Nombre des employés')
plt.title(' Comparaison du satisfaction des employés par Departement')
```

### Test de chideux (chi2)

 H0 = il n'y a pas d'association entre la satisfaction des employés et le departement


 H1 = il y a  une association entre la satisfaction des employés et le departement

```python
# importation du package du test
from scipy.stats import chi2_contingency

# créer des tableaux des contingences
contingency_table = pd.crosstab(df["Department"], df["Satisfaction_Level"])

# Effectuer le test de chi-carréé
chi2_stat, p_value, dof, expected_freq = chi2_contingency (contingency_table)

# Affichage du resultat 
print('statistique de chi_carré :', chi2_stat)
print('P_value :', p_value)
print('Dégrée de liberté :', dof)
print(' Fréquence attendue :\n', expected_freq)
```

p_value est inférieur à 0.05 donc on rejette l'hypothèse nul :
la satisfaction des employés est significativement liée au departement

### Mesure de l'intensité de dépendance 

```python
# Calculer le coefficient de cramer_V
n = contingency_table.sum().sum()
min_din = min(contingency_table.shape)-1
cramer_V = np.sqrt(chi2_stat/(n*min_din))

# Affichage du resultat 
print('La coefficient du cramer_V :', cramer_V)
```

#### Interpretation Finale

La rélation  entre la variable Department et Satisfation_Level est significative mais elle n'est pas forcement important car le coefficient de cramer_V est très proche de 0
