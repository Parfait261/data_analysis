```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
plt.style.use('ggplot')
```

```python
df = pd.read_csv ("data.csv")
df.drop("Hire_Date", axis = 1 , inplace=True )


# Fonction pour créer la colonne Satisfaction_Level
def Add_Satisfaction_Level(score) : 
  if score < 2 :
        return 'Faible'
  elif score < 3.5 :
        return 'moyen'
  elif score >= 3.5 :
      return 'élevé'


#Appel à la fonction 
df["Satisfaction_Level"] = df["Employee_Satisfaction_Score"].apply(Add_Satisfaction_Level)
# Suppression des intrus
df= df[df.Gender != 'Other']
```

```python
# Test de normalité du variable Performance_Score

from scipy.stats  import shapiro
for group in df['Department'].unique() :
    stats, p = shapiro(df[df['Department']== group]['Performance_Score'])
    print(f"{group} : p-value = {p}")
```

Ces deux variables ne suit pas une distribution normales donc il faut utiliser le test Kruskal_Wallis pour comparer la moyenne de score de performance de chaque departement

```python
#Test de Kruskal-Wallis
from scipy.stats import kruskal

# Créer les groupes
groups = [df[df["Department"] == dept]["Performance_Score"] for dept in df["Department"].unique()]

# Test de Kruskal-Wallis
stat, p = kruskal(*groups)

print(f"Statistique de Kruskal-Wallis : {stat:.3f}")
print(f"p-value : {p:.5f}")
```

 Interprétation :
p ≥ 0.05 → pas de différence significative détectée
