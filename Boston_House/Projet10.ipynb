```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('HousingData.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Suppression des valeurs manquantes
df.dropna(inplace = True)
```

```python
#

X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
# Diviser le dataset
X_train , X_test, Y_train, Y_test = train_test_split(X,Y,test_size =0.2)
```

```python
# Construire le modèle
Regressor = LinearRegression()
Regressor.fit(X_train, Y_train)
```

```python
# Faire des prediction 
Y_pred = Regressor.predict(X_test)
Y_pred
```

```python
# Comparaison du resultat 
new_data = pd.DataFrame({'valeur exacte':Y_test ,'valeur predit': Y_pred })
new_data
```

```python
# Calcule des métrique 
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("MSE :", mse)
print("RMSE :", rmse)
print("MAE :", mae)
print("R² :", r2)
```

```python
# Validation croisé 
cv_score = cross_val_score (Regressor, X, Y, scoring='neg_mean_squared_error', cv=5)
cv_rmse = np.sqrt(-cv_score)

print("Cross-validation RMSE par fold :", cv_rmse)
print("RMSE moyen CV :", np.mean(cv_rmse))
```

```python
# les residus 
residus = Y_test - Y_pred

plt.scatter(Y_pred, residus)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Valeurs prédites")
plt.ylabel("Résidus")
plt.title("Analyse des résidus")
plt.show()
```

“Le modèle de régression linéaire utilisé explique environ 78% de la variance des prix des maisons. Les prédictions sont globalement correctes, avec une erreur moyenne d’environ 3–4 unités. Quelques observations présentent des écarts plus élevés. La performance du modèle est stable selon les validations croisées, bien qu’un léger biais apparaisse pour certaines valeurs. Le modèle est adapté pour des estimations globales mais pourrait être amélioré avec d’autres variables ou des techniques plus avancées.”
